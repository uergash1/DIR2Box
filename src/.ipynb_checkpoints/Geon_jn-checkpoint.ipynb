```python
import json
import torch
import numpy as np
import pandas as pd
from tqdm import tqdm, trange
import matplotlib.pyplot as plt
import torch.nn.functional as F
```

```python
#dataset_name = 'clueweb09b_123'
#dataset_name = 'fedweb14'
dataset_name = 'gov2'
```

```python
resource_query_similarity = pd.read_csv(f'../data/{dataset_name}/processed/resource_query_similarity_v3.tsv', sep='\t')
```

```python
resource_query_similarity['query_id'].nunique()
```

```python
scores = list(resource_query_similarity['similarity_score'].values)
```

```python
scores.count(-1), scores.count(0), len(scores) - scores.count(0) - scores.count(-1)
```

```python
query_ids = sorted(list(resource_query_similarity['query_id'].unique()))
queries = torch.from_numpy(np.load(f"../data/{dataset_name}/embeddings/queries.npy"))
```

```python
resource_ids = sorted(list(resource_query_similarity['resource_id'].unique()))
with open(f'../data/{dataset_name}/processed/id_to_rname.json') as f:
    id_to_rname = json.load(f)
        
documents = {}
for resource_id in resource_ids:
    documents[resource_id] = np.load(f"../data/{dataset_name}/embeddings/resources/title_body/{id_to_rname[str(resource_id)]}.npy")
    
resource_document_embedding = torch.Tensor([documents[resource_id] for resource_id in resource_ids])
```

```python
a = torch.mean(resource_document_embedding, 1)
queries.shape
```

```python
A = F.normalize(a, p=2, dim=1)
B = F.normalize(queries, p=2, dim=1)
sim = F.cosine_similarity(A.unsqueeze(1), B.unsqueeze(0), dim=2)
```

```python
plt.imshow(sim.numpy())
plt.colorbar()
plt.show()
```

```python
pos, neg, unk = [], [], []

for resource_id in tqdm(resource_ids):
    for query_id in query_ids:
        score = resource_query_similarity[(resource_query_similarity['query_id']==query_id) & (resource_query_similarity['resource_id']==resource_id)]['similarity_score'].values[0]
        if score > 0:
            pos.append((resource_id, query_id))
        if score == 0: 
            neg.append((resource_id, query_id))
        if score < 0:
            unk.append((resource_id, query_id))
```

```python
len(pos), len(neg), len(unk)
```

```python
pos_sim = [sim[r][q].item() for (r,q) in pos]
neg_sim = [sim[r][q].item() for (r,q) in neg]
unk_sim = [sim[r][q].item() for (r,q) in unk]
```

```python
plt.hist(pos_sim)
plt.show()
```

```python
plt.hist(neg_sim)
plt.show()
```

```python
plt.hist(unk_sim)
plt.show()
```

```python
queries.shape, resource_document_embedding.shape
```

```python
sim = {}
for i in trange(len(queries)):
    sim[i] = {}
    for j in range(len(resource_document_embedding)):
        A = F.normalize(queries[i].unsqueeze(0), p=2, dim=1)
        B = F.normalize(resource_document_embedding[j], p=2, dim=1)
        cos = F.cosine_similarity(A.unsqueeze(1), B.unsqueeze(0), dim=2).view(-1)
        sim[i][j] = cos
```

```python
pos_cnt, neg_cnt, unk_cnt = [], [], []

t = 0.7

for (r,q) in pos:
    pos_cnt.append(torch.sum(sim[q][r] > t).item() / 100)
for (r,q) in neg:
    neg_cnt.append(torch.sum(sim[q][r] > t).item() / 100)
for (r,q) in unk:
    unk_cnt.append(torch.sum(sim[q][r] > t).item() / 100)
```
